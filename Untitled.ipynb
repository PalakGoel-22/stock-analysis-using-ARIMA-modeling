```python
import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller
```

```python
datam.shape
```

```python
train=datam[0:200]
train
```

```python
t=train["Close"].values
result=adfuller(t)
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])

```

```python
#this p value shows that data is non stationary
```

```python
datam['close_log']=np.log(datam['Close'])
datam['close_diff']=datam['close_log']-datam['close_log'].shift(1)
datam['close_diff'].dropna().plot()

result=adfuller(datam['close_diff'].dropna())
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
```

```python
t=train["Open"].values
result1=adfuller(t)
print('ADF Statistic: %f' % result1[0])
print('p-value: %f' % result1[1])
```

```python
datam['open_log']=np.log(datam['Open'])
datam['open_diff']=datam['open_log']-datam['open_log'].shift(1)
datam['open_diff'].dropna().plot()

result2=adfuller(datam['open_diff'].dropna())
print('ADF Statistic: %f' % result2[0])
print('p-value: %f' % result2[1])
```

```python
from statsmodels.tsa.arima_model import ARIMA
step=auto_arima(datam['Close'], trace=True , suppress_warnings= True)
step.summary()
```

```python
model=ARIMA(datam.Close, order=(1,1,0))
result=model.fit(disp=0)
print(result.summary())
```

```python
print(datam.shape)
n=int(len(datam)*0.8)
train=datam.Close[ :n]
test=datam.Close[n: ]
model=ARIMA(train,order=(1,1,0))
result=model.fit(disp=0)
result.summary()
```

```python
end=len(train)+len(test)-1
start=len(train)
pred=model.predict(start=start, end=end, typ='levels')
pred.index=datam.index[start:end+1]
print(pred)
```

```python
step=100
fc,se,conf=result.forecast(step)
```
